Steps to start Zookeeper:

Step - 01: Run ```head -n 10 run-kafka_zookeeper_server.sh``` and see if there are any prerequisites\
Step - 02: If no prerequisites Run ```./run-kafka_zookeeper_server.sh -s start```.\
Step - 03: As there will be block in current terminal(means, won't able to run any new command like in ```./run-hivemetastore.sh -s start```), Open new terminal.\
Step - 04: Check jps, there should be process running "```QuorumPeerMain```".

Steps to start kafka server:

Step - 01: Run ```head -n 10 run-kafka_server.sh```\
Step - 02: Run ```./run-kafka_server.sh -s start```, This will also block terminal. Henc, open new terminal.\
Step - 03: Now we should be able to see running process of "```Jps```", "```Kafka```", "```QuorumPeerMain```".

### TOPIC

Run ```kafka-topics.sh```

Step - 01: Create a topic.

Step - 02: ```kafka-topics.sh --zookeeper 127.0.0.1:2181 --topic first_topic --create```. As we have not given number of partitions, it will show us documentation not error.

Step - 03: Run ```kafka-topics.sh --zookeeper 127.0.0.1:2181 --topic first_topic --create --partitions 3```. It will also show documentation as we need to mention replication-factor

Step - 04: Run ```kafka-topics.sh --zookeeper 127.0.0.1:2181 --topic first_topic --create --partitions 3 --replication-factor 2```. Now, it will give a valid error. Error: "Error while executing topic command : Replication factor: 2 larger than available brokers: 1."

Step - 05: Now change replication-factor to 1, and Run the same command: ```kafka-topics.sh --zookeeper 127.0.0.1:2181 --topic first_topic --create --partitions 3 --replication-factor 1```. Hence, Topic will be created.

Step - 06: Command to see how many topics do we have: ```kafka-topics.sh --zookeeper 127.0.0.1:2181 --list```. It will list all the topics we have created.

Step - 07: Command to Describe the topic: Run ```kafka-topics.sh --zookeeper 127.0.0.1:2181 --describe```. It will describe the topics fully, Topic names, Partition Count, ReplicationFactor, and Configs: TopicName, Partition Number, Leader, replicas, ISR (In-Sync Replicas). Note: Leader, Replicas, ISR: will have ID of broker where they are being assigned to.

Note: ```--delete``` used instead of ```--create```  will delete the speciifc topic.

### PRODUCER

Run ```kafka-console-producer.sh```

Step - 01: Run ```kafka-console-producer.sh --broker-list 127.0.0.1:9092 --topic first_topic``` (Socket: 127.0.0.1:9092) (Socket = Host+Port) Why socket is needed? Bcz A socket is one endpoint of a two-way communication link between two programs running on the network.

Step - 02: Run ```kafka-console-producer.sh --broker-t 127.0.0.1:9092 --topic first_topic --producer-property acks=all``` (acks=all; Leader + ISR acknowledgement)


### CONSUMER

Run ```kafka-console-consumer.sh```

Step - 01: Run ```kafka-console-consumer.sh --bootstrap-server 127.0.0.1:9092 --topic first_topic```.  It would be waiting for messages. So, we need to start producer in one  and consumer in other terminal, which will give Synchronous Communication.

Step - 02: Run ```kafka-console-consumer.sh --bootstrap-server 127.0.0.1:9092 --topic first_topic --from-beginning```. It will consume all the messages from beginning. It's order won't be guaranteed bcz it is dependent on RR Scheduling algo (Premptive Scheduling) and we haven't pass any key also.

### CONSUMER GROUP

Step - 01: Run this command to create consumer group which will consume our messages in particular group only.: ```kafka-console-consumer.sh --bootstrap-server 127.0.0.1:9092 --topic first_topic --group my-first-application```.

There can be a possibility that one consumer group is listening to 2 or 3 partitions. whereas other is listening to maybe one partitions only. Simple Answer is Round Robin Scheduling. If there are 3 Partitions, and 3 Consumer Groups then there would be One-to-One Mapping.

If we will make Consumer Groups > No. of Partitions then as we know that one consumer group will be in inactive state. It will get activated when one group will be crashed. Crashing means press ```Cntrl+C``` in one of the consumer group as of now.
